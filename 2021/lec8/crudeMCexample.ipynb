```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def U(): 
    return np.random.rand()

def Unif(a, b): 
    return (b - a)*U() + a



def f(x,y,z):
    return (5/(648*np.pi))*(x**2 + y**2)


def integral():
    R   = 3
    N   = 100
    M   = 0
    #Vhm = (2*R)**3  # volume of the cube
    V   = (4*np.pi/3.)*R**3. # volume of sphere
    Σf  = 0
    Σf2 = 0
   
    for i in range(N):
        x = Unif(-R, R)
        y = Unif(-R, R)
        z = Unif(-R, R)
        if x**2 + y**2 + z**2 <= R**2:
            M   += 1
            z    = f(x, y, z)
            Σf  += z
            Σf2 += z**2
        

    μf  = Σf/M
    σf2 = Σf2/M - μf**2
    sf2 = (N/(N-1))*σf2
    sf  = sf2**0.5

    I  = V*μf
    δI = V*sf/N**0.5

    #print('I = %2.3f ± %2.3f'%(I, δI))
    return I

        
```

```python
integrals = [integral() for i in range(10000)]
plt.hist(integrals, bins = 50, ec= 'black')
integrals = np.array(integrals)
print('mean I = %2.3f ± %2.3f'%(integrals.mean(), integrals.std()))
```
