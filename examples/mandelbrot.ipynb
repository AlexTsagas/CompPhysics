# Mandelbrot set 

---

code taken from https://docs.scipy.org/doc/numpy/user/quickstart.html see also [Mandelbrot in Wikipedia](https://en.wikipedia.org/wiki/Mandelbrot_set) and closely related [Julia set](https://en.wikipedia.org/wiki/Newton_fractal)


---

```python
import numpy as np
import matplotlib.pyplot as plt
def mandelbrot(h, w, maxit=20 ):
    """Returns an image of the Mandelbrot fractal of size (h,w)."""
    y,x = np.ogrid[ -1.4:1.4:h*1j, -2:0.8:w*1j ]
    c = x+y*1j
    z = c
    divtime = maxit + np.zeros(z.shape, dtype=int)

    for i in range(maxit):
        z = z**2 + c
        diverge = z*np.conj(z) > 2**2         # who is diverging
        div_now = diverge & (divtime==maxit)  # who is diverging now
        divtime[div_now] = i                  # note when
        z[diverge] = 2                        # avoid diverging too much
    return divtime

ax, fig = plt.subplots(figsize=(10,10))
plt.imshow(mandelbrot(400,400, 20))
plt.show()
```

```python
# printing out steps to understand to the code
import numpy as np
import matplotlib.pyplot as plt
def mandelbrotSimple(h =4, w=4, maxit=2 ):
    """Returns an image of the Mandelbrot fractal of size (h,w)."""
    y,x = np.ogrid[ -1.4:1.4:h*1j, -2:0.8:w*1j ]
    c = x+y*1j
    z = c
    print('z = \n', z)
    divtime = maxit + np.zeros(z.shape, dtype=int)
    print('divtime = \n', divtime)

    for i in range(maxit):
        z = z**2 + c
        diverge = z*np.conj(z) > 2**2         # who is diverging
        print('diverge = \n', diverge)
        print('divtime==maxit \n', divtime==maxit)
        div_now = diverge & (divtime==maxit)  # who is diverging now
        print('div_now = diverge & (divtime==maxit) \ n', div_now)
        divtime[div_now] = i                  # note when
        print('divtime \n', divtime)
        z[diverge] = 2                        # avoid diverging too much
        print('z = \n', z)
    return divtime

ax, fig = plt.subplots(figsize=(10,10))
plt.imshow(mandelbrotSimple())
plt.show()
```
